#### Ollama embedding techniques


```python
from langchain.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model="gemma:2b")
)
```

```python
embeddings
```

```python
embedding_docs = embeddings.embed_documents(["Alpha is the first letter of Greek alphabet.", "Beta is the first letter of Greek alphabet."])
```

```python
embedding_docs[1]
```

```python
embeddings.embed_query("what is the second letter of Greek alphabet?")
```
